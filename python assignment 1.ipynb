i) Write a Python program that takes user input for two numbers and performs all arithmetic operations on them. Display the results with appropriate messages for each operation.

```python
# Python program to perform arithmetic operations on two numbers

# Taking user input
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Performing operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Undefined (cannot divide by zero)"
floor_division = num1 // num2 if num2 != 0 else "Undefined (cannot divide by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (cannot divide by zero)"
exponentiation = num1 ** num2

# Displaying results
print("\nResults of Arithmetic Operations:")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
print(f"Floor Division: {num1} // {num2} = {floor_division}")
print(f"Modulus: {num1} % {num2} = {modulus}")
print(f"Exponentiation: {num1} ** {num2} = {exponentiation}")
```

ii) Write a python program to interchange a list items with & without using temporary values.

```python
# Swapping list items using a temporary variable

# Step 1: Create a list
my_list = [10, 20, 30, 40, 50]

print("Original List:", my_list)

# Step 2: Swap first and last elements using a temporary variable
temp = my_list[0]        # store first element in temp
my_list[0] = my_list[-1] # put last element in first place
my_list[-1] = temp       # put temp (old first) in last place

print("List after swapping (using temp):", my_list)
```

iii) Create a dictionary, do all operations like addition of elements, remove elements, copy item & create nested dictionaries too.

```python
# Step 1: Create a dictionary
student = {
    "name": "Aman",
    "age": 20,
    "course": "Pharmacy"
}
print("Original Dictionary:", student)

# Step 2: Add new elements
student["city"] = "Delhi"        # adding new key-value pair
student["grade"] = "A"
print("After Adding Elements:", student)

# Step 3: Remove elements
student.pop("age")               # removes a specific key
print("After Removing 'age':", student)

del student["grade"]             # deletes another key
print("After Deleting 'grade':", student)

# Step 4: Copy dictionary
student_copy = student.copy()
print("Copied Dictionary:", student_copy)

# Step 5: Nested Dictionary
students = {
    "student1": {"name": "Aman", "age": 20},
    "student2": {"name": "Ravi", "age": 22},
    "student3": {"name": "Priya", "age": 21}
}
print("Nested Dictionary of Students:", students)
```
